### Basic Probability

##### <div class="alert alert-block alert-success"> Probability of an Event </div>

* A __random phenomenon__ is an observable situation for which the outcome is uncertain. 

* The __sample space__ is the set of all possible outcomes for a random phenomenon. 

* An __event__ is a subset of the sample space and corresponds to a particular outcome or group of possible outcomes. 

* The __probability__ of an event is the proportion of times that an event would occur in a long run of observations. 

For the probability of event A: <br>
1. $0\leq P(A)\leq 1$ <br>
2. If $P(A)=1$, then $A$ is guaranteed to occur.<br>
3. If $P(A)=0$, then $A$ is guaranteed to not occur. <br>
4. The sum of the probabilities of all outcomes in the sample space is 1. 


* When all outcomes are __equally likely__, then the theoretical probability of event $A$ is given by:
$$P(A)=\frac{\text{number of outcomes in event $A$}}{\text{number of outcomes in the sample space}}$$

* The __complement__ of an event $A$ is the set of all outcomes in the sample space that are NOT in $A$. It is denoted by $A^c$.

$$P(A^c)=1-P(A)$$

* The __conditional probability__ of $B$, given $A$ has occured is $P(B|A).$
$$P(B|A)=\frac{P(A \text{ and } B)}{P(A)}$$

* Rearranging we obtain the __general multiplication rule__: 
$$P(A \text{ and } B)=P(A)\times P(B|A)$$


* Two events are __independent__ if the probability that one occurs is not affected by whether the other event occurs. That is, if $P(A|B)=P(A),$ then $A$ and $B$ are independent.
Note: Similarly, if $P(B|A)=P(B),$ then $A$ and $B$ are independent. 

Note: In either case, the multiplication rule simplies to $P(A \text{ and } B)=P(A)\times P(B)$.

* The __general addition rule__ is given by: 
$$P(A \text{ or } B)= P(A)+P(B)-P(A \text{ and } B)$$

* Two events are __disjoint__ if they cannot both occur. That is, $P(A \text{ and } B)=0.$ Note: For disjoint events, the addition rule simplifies to $P(A \text{ or } B)= P(A)+P(B)$.

Events cannot be both disjoint and independent (being disjoint is a type of dependency).

### Discrete Distributions

The probability distribution must satisfy the following:
1. For each $x$ (value of the random variable), $0 \leq P(x) \leq 1.$
2. $\Sigma P(x) =1$


Note: Since probability distributions include all values of a random variable, we'll use population parameters to describe them.

```r
# Figure: a simple discrete probability distribution
x_values <- c(1, 2, 3, 4, 5)
p_values <- c(0.1, 0.2, 0.3, 0.15, 0.25)

# Create a bar plot for the discrete distribution
barplot(p_values, names.arg = x_values, col = "lightblue",
        main = "Discrete Probability Distribution",
        xlab = "x (Value of Random Variable)", 
        ylab = "P(x) (Probability)",
        ylim = c(0, 1))

# Add grid lines
grid(nx = NA, ny = NULL)
```

##### <div class="alert alert-block alert-success"> Binomial Distribution </div>

A binomial experiment with $n$ identical trials satisfies the following:
1. Each trial has 2 possible outcomes: success and failure.
2. Each trial has the same probability of success. $P$(success)$=p$ and $P$(failure)$=1-p$. 
3. The $n$ trials are independent. 

Goal: Find the probability of $x$ successes in $n$ trials.

##### <div class="alert alert-block alert-success"> Normal Distribution </div>

* A __normal distribution__ is a bell-shaped curve with its peak above the mean.  

Properties:
1. Bell-shaped and symmetric about its mean.
2. Approaches the horizontal axis but never touches.
3. Area under the curve is equal to 1. 


* The __standard normal distribution__ is the normal distribution with mean $\mu=0$ and standard deviation $\sigma=1.$ 
<br>


* __Empirical Rule__: 
For bell-shaped distributions, <br>
68% of the data falls within 1 standard deviation of the mean <br>
95% of the data falls within 2 standard deviations of the mean<br>
99.7% of the data falls within 3 standard deviations of the mean <br> <br>



```r
# Figure: normal distribution
mean <- 0
sd <- 1

# Create a sequence of x values from -4 to 4 (covering ~99.7% of the data)
x_values <- seq(-4, 4, length = 1000)

# Calculate the normal distribution (density) for each x value
y_values <- dnorm(x_values, mean = mean, sd = sd)

# Plot the normal distribution curve
plot(x_values, y_values, type = "l", lwd = 2, col = "blue",
     xlab = "X values (Standard Deviations)", ylab = "Density",
     main = "Empirical Rule for Normal Distribution", xaxt = "n")

# Manually add x-axis with custom sigma labels (-3σ, -2σ, ..., 3σ)
axis(1, at = c(-3, -2, -1, 0, 1, 2, 3), labels = c("-3", "-2", "-1", "0", "1", "2", "3"))

# Add shading for 68% (1 SD from the mean)
x_shade_68 <- seq(-1, 1, length = 500)
y_shade_68 <- dnorm(x_shade_68, mean = mean, sd = sd)
polygon(c(-1, x_shade_68, 1), c(0, y_shade_68, 0), col = rgb(0.1, 0.1, 0.9, 0.1), border = NA)

# Add shading for 95% (2 SDs from the mean)
x_shade_95 <- seq(-2, 2, length = 500)
y_shade_95 <- dnorm(x_shade_95, mean = mean, sd = sd)
polygon(c(-2, x_shade_95, 2), c(0, y_shade_95, 0), col = rgb(0.1, 0.9, 0.1, 0.1), border = NA)

# Add shading for 99.7% (3 SDs from the mean)
x_shade_997 <- seq(-3, 3, length = 500)
y_shade_997 <- dnorm(x_shade_997, mean = mean, sd = sd)
polygon(c(-3, x_shade_997, 3), c(0, y_shade_997, 0), col = rgb(0.9, 0.1, 0.1, 0.1), border = NA)

# Add vertical lines for mean, +/- 1 SD, 2 SD, 3 SD
abline(v = c(-1, 1), col = "darkblue", lty = 2, lwd = 2)
abline(v = c(-2, 2), col = "darkgreen", lty = 2, lwd = 2)
abline(v = c(-3, 3), col = "darkred", lty = 2, lwd = 2)
abline(v = 0, col = "black", lwd = 2)

# Add arrows to indicate ranges
arrows(-1, 0.1, 1, 0.1, length = 0.1, col = "darkblue", lwd = 2)
arrows(-2, 0.05, 2, 0.05, length = 0.1, col = "darkgreen", lwd = 2)
arrows(-3, 0.005, 3, 0.005, length = 0.1, col = "darkred", lwd = 2)

# Add text labels to explain the empirical rule
text(0, 0.12, "68%", col = "blue")
text(0, 0.07, "95%", col = "darkgreen")
text(0, 0.025, "99.7%", col = "darkred")
```

* The __p$^{th}$ percentile__ of a distribution is the value such that $p$% of the observations are less than or equal to the value. 


```r
# Figure: pth percentile
mean_value <- 15
sd_value <- 2  # Example standard deviation
percentile_value <- 0.35  # 35th percentile

# Create a sequence of x values
x_values <- seq(mean_value - 4 * sd_value, mean_value + 4 * sd_value, length = 1000)

# Calculate the normal distribution (density) for each x value
y_values <- dnorm(x_values, mean = mean_value, sd = sd_value)

# Calculate the x value for the 35th percentile
x_percentile_35 <- qnorm(percentile_value, mean = mean_value, sd = sd_value)

# Plot the normal distribution curve
plot(x_values, y_values, type = "l", lwd = 2, col = "blue",
     xlab = "X values", ylab = "Density",
     main = "Normal Distribution with Mean 15 and 35th Percentile")

# Shade the area under the curve up to the 35th percentile
x_shade <- seq(mean_value - 4 * sd_value, x_percentile_35, length = 500)
y_shade <- dnorm(x_shade, mean = mean_value, sd = sd_value)
polygon(c(mean_value - 4 * sd_value, x_shade, x_percentile_35), 
        c(0, y_shade, 0), col = rgb(0.1, 0.9, 0.1, 0.3), border = NA)

# Add a vertical line at the 35th percentile
abline(v = x_percentile_35, col = "red", lwd = 2, lty = 2)

# Add text label for the 35th percentile
text(x_percentile_35, 0.02, paste("35th Percentile =", round(x_percentile_35, 2)), pos = 4, col = "red")
```

##### <div class="alert alert-block alert-success"> $t$-Distribution </div>

* A __$t$-distribution__ is a bell-shaped curve that has slightly thicker tails (exhibits more variability) than a normal distribution. 


Properties:
1. Bell-shaped and symmetric about 0. 
2. Approaches the horizontal axis but never touches.
3. Area under the curve is equal to 1.
4. Standard deviation is slightly larger than 1 and depends on the degrees of freedom ($df=n-1$) which dictates the shape of the distribution.
5. Approaches the standard normal distribution as the degrees of freedom increase ($df=30$ or more, the distributions are nearly identical). 


```r
# Figure: Normal vs T
x_values <- seq(-4, 4, length = 1000)

# Calculate the normal distribution (mean = 0, sd = 1)
normal_dist <- dnorm(x_values, mean = 0, sd = 1)

# Calculate t-distributions for different degrees of freedom (df)
t_dist_df_2 <- dt(x_values, df = 2)   # t-distribution with df = 2
t_dist_df_5 <- dt(x_values, df = 5)   # t-distribution with df = 5
t_dist_df_30 <- dt(x_values, df = 30) # t-distribution with df = 30

# Plot the normal distribution
plot(x_values, normal_dist, type = "l", lwd = 2, col = "blue", 
     xlab = "X values", ylab = "Density",
     main = "Comparing t-Distributions and Normal Distribution")

# Add the t-distributions to the plot
lines(x_values, t_dist_df_2, col = "red", lwd = 2, lty = 2)   # df = 2
lines(x_values, t_dist_df_5, col = "green", lwd = 2, lty = 3) # df = 5
lines(x_values, t_dist_df_30, col = "purple", lwd = 2, lty = 4) # df = 30

# Add a legend to explain the plot
legend("topright", legend = c("Normal Distribution", "t-distribution df=2", 
                              "t-distribution df=5", "t-distribution df=30"), 
       col = c("blue", "red", "green", "purple"), lwd = 2, lty = c(1, 2, 3, 4))
```

* The __expected frequencies__ $(E)$ in a two-way table are the frequencies we would expect by chance if there were no relationship between the row and the column variables. We use this formula to calculate the expected frequency, E, for each cell:
 $$E =\frac{(row total)(column total)}{
grand total}$$

* The __observed frequencies__ $(O)$ in a two-way table are the frequencies we actually see occur. 

* The __chi-squared statistic__ is given by $$\chi ^ 2=\Sigma \frac{(O-E)^2}{E}$$
Note: The larger the value of $\chi ^ 2$, the greater the average difference between the observed and expected
frequencies in the cells.

* For two-way tables, __degrees of freedom__ are calculated by $df=(r-1)(c-1)$ where $r$ is number of rows and $c$ is number of columns. 

Properties for $\chi ^2$ distribution:
1. The range is $[0, \infty)$.
2. Mean of distribution is $df$ and standard deviation is $\sqrt{2df}$.
3. Right-skewed distribution but as the $df$ increases, the skew lessens and the curve becomes more bell-shaped.
4. Area under the curve is equal to 1.

The __Sampling Distribution of the Sample Mean__ has: 
$\mu_{\bar{x}}=\mu$ and $\sigma_{\bar{x}}=\frac{\sigma}{\sqrt{n}}$. (These are the mean and standard deviation of the sample means).

Thus, as $n$ increases, the variability of the sample means decreases (because of the way standard deviation is calculated).

__Central Limit Theorem__: For a population with any distribution, the sampling distribution of the sample mean approaches an approximately normal distribution as $n$ increase. It can be considered a normal distribution for $n\geq 30.$ 

Note: If the original distribution for a population is normal, then the sampling distribution of the sample mean will be normal for $any$ sample size.

Recall: A sample statistic is a random variable which means it can take on different values so we can calculate the probability of it being a certain value. 

When dealing with categorical variables, we calculate proportions instead of means. So we want to look at the possible sample proportions and their corresponding probabilities. 

The __Sampling Distribution of the Sample Proportion__ has: 
$\mu_{\hat{p}}=p$ and $\sigma_{\hat{p}}=\sqrt{\frac{p(1-p)}{n}}$

The sampling distrubution of the sample proportion is approximately normal if: 
$$np\geq 15 \text{ and } n(1-p)\geq 15 \text{ and } n < (10\% \times N)$$
